```python
from src.dwt.dwt import Dwt, Dwt_custom
from src.dwt.wavelets import Haar
import torch
import matplotlib.pyplot as plt
from data_loader import ISIC
from torch.utils.data import DataLoader
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

bdir = "../../../../../pscratch/sd/m/mati/half_yuki_sim_64"
file = "data.mdb"
transformer1 = None
dataset = ISIC(bdir, file, transformer1, 1, False)
loader = DataLoader(dataset, batch_size=32, shuffle=False, pin_memory=True)
dataloader_iterator = iter(loader)
x = dataloader_iterator.next().to(device)

wavelet = Haar().to(device)
dwt = Dwt(wavelet=wavelet).to(device)
dwt_custom = Dwt_custom(wavelet=wavelet).to(device)


y = dwt.forward(x)

y2 = dwt_custom.forward(x)

x_reconstructed = dwt.inverse(y)

# Calculate reconstruction error
error = torch.mean((x - x_reconstructed) ** 2).item()
exact_match = torch.allclose(x, x_reconstructed, atol=1e-6)

print("Reconstruction error:", error)
print("Exact match:", exact_match)


error = torch.mean((y['low'] - y2['low']) ** 2).item()
exact_match = torch.allclose(y['low'], y2['low'], atol=1e-6)

print("Reconstruction error low:", error)
print("Exact match low:", exact_match)    

error = torch.mean((y['high'][:, [3, 4, 5], :, :] - y2['high'][:, 3:6, :, :]) ** 2).item()
exact_match = torch.allclose(y['high'], y2['high'], atol=1e-6)

print("Reconstruction error high:", error)
print("Exact match low:", exact_match)  




```

```python
def plot_images(original, reconstructed):
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    original = original.permute(1, 2, 0)
    axs[0].imshow(original.cpu().numpy())
    axs[0].set_title("Original Image")
    axs[1].imshow(reconstructed.permute(1, 2, 0).cpu().numpy())
    axs[1].set_title("Reconstructed Image")
    plt.show()
    
def plot_colormap_of_component(component, cmap='viridis'):
    # Squeeze the batch and channel dimensions for the image plot
    component_squeezed = component.squeeze().cpu().detach().numpy()

    # Plot
    plt.figure(figsize=(6, 6))
    plt.contourf(component_squeezed, cmap=cmap)
    plt.colorbar()  # Add a colorbar to show the scale
    plt.title('Colormap of Component')
    plt.show()

# Assuming x and reconstructed_x are normalized image tensors
plot_images(y['high'][:, :3, :, :][0], y2['high'][:, 0:3, :, :][0])

plot_images(y['high'][:, 3:6, :, :][0], y2['high'][:, 3:6, :, :][0])

plot_images(y['high'][:, 6:9, :, :][0], y2['high'][:, 6:9, :, :][0])

plot_colormap_of_component(x[0, 0, :, :])
plot_colormap_of_component(x_reconstructed[0, 0, :, :])
```
